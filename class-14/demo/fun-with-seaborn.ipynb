```python
flights = sns.load_dataset('flights')
flights
```

```python
sns.get_dataset_names()
```

```python
sns.relplot(x='year', y='passengers', hue='month', data=flights)
```

```python
year_sum = flights.groupby('year').sum()
year_sum
```
